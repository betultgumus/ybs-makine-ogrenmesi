>
> **4b)** Her kategorik değişken için kaç benzersiz değer olduğunu `nunique()` ile bulun ve sonuçları ekrana yazdırın.

```python
# ============================================================
# ADIM 4 — Değişken Sözlüğü ve Sınıflandırma
# ============================================================

# 4a) Sayısal ve kategorik sütunları ayrı listeler olarak yazdırın
# Kodu buraya yazınız
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Sayısal sütunlar:", numeric_cols)
print("Kategorik sütunlar:", categorical_cols, "\n")


# 4b) Her kategorik değişken için benzersiz değer sayısını (nunique) bulun
# Kodu buraya yazınız
for col in categorical_cols:
    unique_count = df[col].nunique()
    print(f"{col} sütununda benzersiz değer sayısı: {unique_count}")


# 4c) Değişken türlerini bar grafiğiyle görselleştirin
target_col = ['Churn']
cat_cols = [c for c in categorical_cols if c not in target_col]

dtype_counts = pd.Series({
    'Sayısal': len(numeric_cols),
    'Kategorik': len(cat_cols),
    'Hedef': len(target_col)
})

fig, ax = plt.subplots(figsize=(6, 4))
dtype_counts.plot(kind='bar', color=sns.color_palette('pastel', n_colors=3), edgecolor='black', ax=ax)
ax.set_title('Değişken Türlerinin Dağılımı')
ax.set_ylabel('Sütun Sayısı')
ax.set_xlabel('Değişken Türü')
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

for i, v in enumerate(dtype_counts):
    ax.text(i, v + 0.2, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

> **5d)** `seaborn.heatmap(df.isnull())` ile eksik veri ısı haritasını görselleştirin. Temizleme öncesi ve sonrasını karşılaştırın.

```python
# ============================================================
# ADIM 5 — Eksik Veri Analizi ve Temizleme
# ============================================================

# 5a) Sütun bazlı eksik değer sayısı ve yüzdesi
missing_df = pd.DataFrame({
    'Sütun': df.columns,
    'Eksik Değer Sayısı': df.isnull().sum().values,
    'Eksik Değer Yüzdesi': (df.isnull().sum().values / len(df) * 100).round(2)
})

# Eksik değer sayısına göre azalan sırada sırala
missing_df = missing_df.sort_values('Eksik Değer Sayısı', ascending=False).reset_index(drop=True)

# Eksik değerleri olan satırları göster

if missing_df['Eksik Değer Sayısı'].max() == 0:
    print("Veri setinde eksik değer yoktur.\n")
else:
    print(missing_df[missing_df['Eksik Değer Sayısı'] > 0])

#display(missing_df)


# 5b) TotalCharges'taki gizli boşluk karakterli satırları tespit edin
# Kodu buraya yazınız
whitespace_mask = df['TotalCharges'].astype(str).str.strip() == ''
whitespace_rows = df[whitespace_mask]

print(f"Boşluk karakterli satır sayısı: {len(whitespace_rows)}")
if len(whitespace_rows) > 0:
    print("\nBoşluk karakterli satırlar:")
    display(whitespace_rows[['customerID', 'TotalCharges']])
else:
    print("TotalCharges sütununda boşluk karakterli satır yoktur.")



# 5c) TotalCharges'ı float64'e çevirin, NaN satırlarını kaldırın
# Kodu buraya yazınız
print(f"\nTemizleme öncesi şekil: {df.shape}")

# TotalCharges'ı float64'e çevir (boşluk karakterlerini NaN'a çevir)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# NaN satırlarını kaldır
df = df.dropna(subset=['TotalCharges'])
# Boşluk karakterli olanları
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan).astype(float)

print(f"Temizleme sonrası şekil: {df.shape}")
print(f"\nTotalCharges veri tipi: {df['TotalCharges'].dtype}")



# 5d) Eksik veri ısı haritasını görselleştirin (seaborn heatmap)
# Kodu buraya yazınız

if missing_df['Eksik Değer Sayısı'].max() == 0:
    print("\nVeri setinde eksik değer olmadığından ısı haritası oluşturulamaz.")
else:
    plt.figure(figsize=(14, 8))
    sns.heatmap(df.isnull(), yticklabels=False, cbar=True, cmap='viridis')
    plt.title('Eksik Veri Isı Haritası', fontsize=16, fontweight='bold', pad=15)
    plt.xlabel('Sütunlar', fontsize=12)
    plt.ylabel('Satırlar (Müşteriler)', fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
# ============================================================
# ADIM 7 — Sayısal Değişkenlerin Dağılım Analizi
# ============================================================

# 7a) tenure, MonthlyCharges, TotalCharges için istatistiksel özet
# Kodu buraya yazınız
display(df[['tenure', 'MonthlyCharges', 'TotalCharges']].describe())



# 7b) Her sayısal değişken için histogram + KDE grafiği (subplot)

# Sayısal değişkenler
numeric_vars = ['tenure', 'MonthlyCharges', 'TotalCharges']

# Figür ve subplotlar (1x3)
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Sayısal Değişkenlerin Dağılımı (Histogram + KDE)', fontsize=16, fontweight='bold', y=1.02)

# Her değişken için histogram + KDE çiz
for idx, col in enumerate(numeric_vars):
    sns.histplot(data=df, x=col, kde=True, bins=30, color='steelblue', alpha=0.6, ax=axes[idx])
    axes[idx].set_title(f'{col}', fontsize=13, fontweight='bold')
    axes[idx].set_xlabel('Değer', fontsize=11)
    axes[idx].set_ylabel('Frekans', fontsize=11)
    axes[idx].grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()

print('''
1. tenure (Müşteri Bağlılık Süresi) -> Bimodal 

Yorum: Veriler uç noktalarda yoğunlaşmış durumda. Sol taraftaki (0-5 ay arası) yoğunluk, çok yeni müşterilerin fazla olduğunu; sağ taraftaki (70+ ay) yoğunluk ise sadık, uzun süreli müşterilerin de ciddi bir ağırlığı olduğunu gösteriyor.
İstatistiksel Durum: Normal dağılımdan oldukça uzaktır. Orta kısımlar daha düz (uniform) bir yapıya yakındır.

---

2. MonthlyCharges (Aylık Ücretler) -> Bimodal 
Yorum: En büyük yığılma 20 birim civarındaki düşük ücretli paketlerde. Ancak 70 ile 110 birim arasında da geniş bir "tepe" (kitle) bulunuyor. Bu, muhtemelen temel hizmetleri alan büyük bir grup ile yüksek fiyatlı (premium/ek paketli) hizmetleri alan ikinci bir ana grubun varlığına işaret eder.
İstatistiksel Durum: Çok tepeli (multimodal) bir yapıya yakındır, normal dağılım göstermez.

---

3. TotalCharges (Toplam Ücretler) -> Pozitif Çarpık (Right-Skewed)

Yorum: Verilerin çoğu sol tarafta (düşük değerlerde) toplanmış, sağa doğru ise bir "kuyruk" uzanıyor. Bu çok doğaldır; çünkü çoğu müşteri ya yenidir (tenure düşük) ya da düşük ücretli paket kullanıyordur. Yüksek toplam harcamaya ulaşan müşteri sayısı doğal olarak daha azdır.
İstatistiksel Durum: Çarpıklık (skewness) değeri yüksektir. Mod < Medyan < Ortalama ilişkisi görülür.

''')



# 7c) Her sayısal değişken için box plot
# Figür ve subplotlar (1x3) - Box plot
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Sayısal Değişkenlerin Box Plot Analizi (Aykırı Değer Tespiti)', 
             fontsize=16, fontweight='bold', y=1.02)

# Her değişken için boxplot çiz
for idx, col in enumerate(numeric_vars):
    sns.boxplot(data=df, y=col, color='lightblue', ax=axes[idx])
    axes[idx].set_title(f'{col} - Box Plot', fontsize=13, fontweight='bold')
    axes[idx].set_ylabel('Değer', fontsize=11)
    axes[idx].grid(axis='y', alpha=0.3, linestyle='--')
    
    # İstatistikler
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    
    axes[idx].text(0.02, 0.98, f'Aykırı Değer Sayısı: {len(outliers)}', 
                   transform=axes[idx].transAxes, fontsize=10, verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))

plt.tight_layout()
plt.show()

# Box plot analizi yorumu
print("\n" + "="*80)
print("7C) BOX PLOT ANALİZİ VE AYKIRIDEGER TESPİTİ")
print("="*80)

print(f"""
📊 Box Plot Nedir?
   • Q1 (25. yüzdelik) - alt çizgi
   • Median (50. yüzdelik) - ortadaki çizgi
   • Q3 (75. yüzdelik) - üst çizgi
   • IQR = Q3 - Q1 (merkezi %50)
   • Aykırı değerler: < Q1-1.5×IQR veya > Q3+1.5×IQR

🔍 Aykırı Değer Analizi (Outliers):
""")

for col in numeric_vars:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    outlier_pct = (len(outliers) / len(df)) * 100
    
    print(f"\n   {col}:")
    print(f"   • Q1: {q1:.2f}, Q3: {q3:.2f}, IQR: {iqr:.2f}")
    print(f"   • Aykırı Değer Sınırları: [{lower_bound:.2f}, {upper_bound:.2f}]")
    print(f"   • Aykırı Değer Sayısı: {len(outliers)} ({outlier_pct:.2f}%)")
    if len(outliers) > 0:
        print(f"   • Min Aykırı: {outliers.min():.2f}, Max Aykırı: {outliers.max():.2f}")




# 7d) Çarpıklık (skewness) ve basıklık (kurtosis) hesabı
print("\n" + "="*90)
print("7D) ÇARPIKLIK (SKEWNESS) VE BASIKLIK (KURTOSIS) - DETAYLI ANALİZ")
print("="*90)

# Skewness ve Kurtosis hesapla
skewness = df[numeric_vars].skew()
kurtosis = df[numeric_vars].kurtosis()

print("\n📊 TEMEL İSTATİSTİKLER:")
print("-" * 90)

# Özet tablosu
stats_summary = pd.DataFrame({
    'Değişken': numeric_vars,
    'Mean': [f"{df[col].mean():.2f}" for col in numeric_vars],
    'Median': [f"{df[col].median():.2f}" for col in numeric_vars],
    'Std Dev': [f"{df[col].std():.2f}" for col in numeric_vars],
    'Min': [f"{df[col].min():.2f}" for col in numeric_vars],
    'Max': [f"{df[col].max():.2f}" for col in numeric_vars],
    'Skewness': [f"{skewness[col]:.4f}" for col in numeric_vars],
    'Kurtosis': [f"{kurtosis[col]:.4f}" for col in numeric_vars]
})

print(stats_summary.to_string(index=False))

print("\n\n" + "="*90)
print("📈 ÇARPIKLIK (SKEWNESS) YORUMU - Her Değişken İçin Detaylı Analiz")
print("="*90)

print("""
🎯 SKEWNESS AYARLAMA ÖLÇEĞI:
   • Değer = 0       → Mükemmelen simetrik dağılım (Normal)
   • 0 < |Skew| < 0.5  → Hafif çarpık (Neredeyse simetrik)
   • 0.5 < |Skew| < 1  → Orta derecede çarpık
   • |Skew| > 1     → Şiddetli çarpık

🔸 Pozitif Skewness (Sağa Çarpık):
   → Dağılımın sağ kuyruğu (yüksek değerleri) daha uzun
   → Mean > Median şeklinde gözlenir
   → Yüksek aykırı değerler mevcuttur

🔸 Negatif Skewness (Sola Çarpık):
   → Dağılımın sol kuyruğu (düşük değerleri) daha uzun
   → Mean < Median şeklinde gözlenir
   → Düşük aykırı değerler mevcuttur
""")

for col in numeric_vars:
    skew_val = skewness[col]
    mean_val = df[col].mean()
    median_val = df[col].median()
    
    print(f"\n{'─'*90}")
    print(f"✓ {col.upper()}: Skewness = {skew_val:.4f}")
    print(f"{'─'*90}")
    
    # Skewness Kategorisi
    if skew_val > 0.5:
        category = "🔴 SAĞA ÇARPIK (Right-skewed)"
        implication = "Yüksek değerler yer yer vardır, aykırı değerler yukarıda"
    elif skew_val < -0.5:
        category = "🔵 SOLA ÇARPIK (Left-skewed)"
        implication = "Düşük değerler yer yer vardır, aykırı değerler aşağıda"
    else:
        category = "🟢 YAKLAŞIK SİMETRİK"
        implication = "Dağılım dengeli, neredeyse normal dağılıma benzer"
    
    print(f"  Kategori: {category}")
    print(f"  Anlamı: {implication}")
    print(f"  Mean: {mean_val:.2f} | Median: {median_val:.2f}", end="")
    
    if mean_val > median_val:
        print(f" → Mean > Median (Sağa çarpık işareti)")
    elif mean_val < median_val:
        print(f" → Mean < Median (Sola çarpık işareti)")
    else:
        print(f" → Mean ≈ Median (Simetrik dağılım)")
    
    # Özel yorumlar
    if col == 'tenure':
        print(f"  💡 Özel Yorum: Bimodal dağılım (yeni vs eski müşteriler)")
        print(f"     → Düşük tenure (0-6 ay) ve yüksek tenure (60-72 ay) grupları var")
        print(f"     → Müşteri segmentasyonu için uygun")
    elif col == 'MonthlyCharges':
        print(f"  💡 Özel Yorum: Aylık ücretlerde düşük değerler ağırlıklı")
        print(f"     → Düşük maliyetli paketler daha popüler")
        print(f"     → Yüksek ücretli müşteriler azınlık ama önemli")
    elif col == 'TotalCharges':
        print(f"  💡 Özel Yorum: tenure × MonthlyCharges olduğu için daha çarpık")
        print(f"     → Yeni müşteriler düşük toplam ücrete sahip (kısa süre)")
        print(f"     → Eski müşteriler yüksek toplam ücrete sahip (kumülatif)")
        print(f"     → Log transformasyonu uygulanabilir")

print("\n\n" + "="*90)
print("📐 BASIKLIK (KURTOSIS) YORUMU")
print("="*90)

print("""
🎯 KURTOSIS AYARLAMA ÖLÇEĞI (Excess Kurtosis):
   • Değer ≈ 0       → Normal dağılıma benzer (Mesokurtik)
   • Değer > 0       → Daha keskin tepeli, daha ağır kuyruklı (Leptokurtik)
   • Değer < 0       → Daha düz tepeli, daha hafif kuyruklı (Platykurtik)

🔸 Leptokurtik (Excess Kurtosis > 0):
   → Dağılım daha keskin bir tepeye sahip
   → Aykırı değerler daha fazla (ağır kuyruklar)
   → Verinin merkezde daha yoğun olması

🔸 Platykurtik (Excess Kurtosis < 0):
   → Dağılım daha düz bir tepeli
   → Aykırı değerler daha az (hafif kuyruklar)
   → Verinin daha geniş bir aralıkta dağılması
""")

for col in numeric_vars:
    kurt_val = kurtosis[col]
    
    print(f"\n{'─'*90}")
    print(f"✓ {col.upper()}: Excess Kurtosis = {kurt_val:.4f}")
    print(f"{'─'*90}")
    
    # Kurtosis Kategorisi
    if kurt_val > 1:
        category = "🔴 LEPTOKURTIK (Keskin tepeli)"
        implication = f"Verinin merkezde çok yoğun, kuyruklarda birçok aykırı değer"
    elif kurt_val < -0.5:
        category = "🔵 PLATYKURTIK (Düz tepeli)"
        implication = f"Verinin geniş aralıkta dağılımı, az aykırı değer"
    else:
        category = "🟢 MESOKURTIK (Normal benzer)"
        implication = f"Normal dağılıma yakın kurtosis değeri"
    
    print(f"  Kategori: {category}")
    print(f"  Anlamı: {implication}")
    
    # Aykırı değer ilişkisi
    outlier_count = len(df[(df[col] < df[col].quantile(0.25) - 1.5*(df[col].quantile(0.75)-df[col].quantile(0.25))) | 
                            (df[col] > df[col].quantile(0.75) + 1.5*(df[col].quantile(0.75)-df[col].quantile(0.25)))])
    outlier_pct = (outlier_count / len(df)) * 100
    print(f"  Aykırı Değer: {outlier_count} ({outlier_pct:.2f}%)")

print("\n\n" + "="*90)
print("🎯 ML MODELLEMESİ İÇİN ÖNERİLER")
print("="*90)

print("""
1️⃣  ÇARPIK DAĞILIMLAR (Skewness):
   ✓ Power Transformation: Yeo-Johnson, Box-Cox (skewness < -1 veya > 1 için)
   ✓ Log Transformation: TotalCharges gibi sağa çarpık veriler için
   ✓ Robust Scaling: StandardScaler yerine RobustScaler kullanın
   ✓ Tree Models: Gradient Boosting, Random Forest çarpıklığa duyarlı DEĞİLDİR
   ✗ Linear Models: Logistic Regression çarpık verilerde düşük performans gösterir

2️⃣  AYKIRI DEĞERLER (Kurtosis):
   ✓ Isolation Forest: Aykırı değer tespiti ve çıkarması
   ✓ IQR Method: Q1 ± 1.5×IQR dışındaki değerleri çıkarmak
   ✓ Winsorization: Aykırı değerleri sınırlandırmak (clip)
   ✓ Robust Scaling: Aykırı değerlerin etkisini azaltmak

3️⃣  BİMODAL DAĞILIM (tenure):
   ✓ Feature Engineering: Tenure kategorilerine ayır (yeni, aktif, eski)
   ✓ Segmentation: Müşteri segmentine göre ayrı modeller
   ✓ Interaction Features: tenure × churn ilişkisi için özel feature
""")

print("\n" + "="*90 + "\n")

```

```python
# ============================================================
# ADIM 9 — Korelasyon Analizi
# ============================================================

# 9a) Sayısal değişkenler için korelasyon matrisi
#     (tenure, MonthlyCharges, TotalCharges, SeniorCitizen)
numeric_corr_vars = ['tenure', 'MonthlyCharges', 'TotalCharges', 'SeniorCitizen']

# Korelasyon matrisi hesapla (Pearson)
correlation_matrix = df[numeric_corr_vars].corr()

print("\n" + "="*100)
print("9A) SAYISAL DEĞİŞKENLER İÇİN KORELASYON MATRİSİ (Pearson Correlation)")
print("="*100)

print("\n📊 KORELASYON MATRİSİ (Correlation Matrix):")
print("-" * 100)
print(correlation_matrix.to_string())


# 9b) Seaborn heatmap ile korelasyon görselleştirme
#     (annot=True, cmap='coolwarm', fmt='.2f')

fig, ax = plt.subplots(figsize=(10, 8))

# Korelasyon heatmap'ini çiz
sns.heatmap(correlation_matrix, 
            annot=True,           # Her hücrede değer göster
            cmap='coolwarm',      # Renk paleti: mavi(-1) → beyaz(0) → kırmızı(+1)
            fmt='.2f',            # 2 ondalak basamağa kadar göster
            center=0,             # Ortada 0 değeri olacak şekilde
            square=True,          # Hücreler kare olacak
            linewidths=1,         # Hücreler arasında çizgi
            cbar_kws={'label': 'Korelasyon Katsayısı (r)'},
            vmin=-1, vmax=1,      # Renk skalasının sınırları
            ax=ax)

ax.set_title('Sayısal Değişkenler Arasında Korelasyon Matrysi (Heatmap)\ntenure, MonthlyCharges, TotalCharges, SeniorCitizen', 
             fontsize=14, fontweight='bold', pad=20)

# X ve Y etiketlerini sağa/yukarıya kaydır
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.show()

# Heatmap yorumu
print("9B) KORELASYON HEATMAP YORUMU")

print('''Değişkenler Arası Korelasyon Özeti

tenure & TotalCharges (0.83): Beklenen güçlü pozitif ilişki. 
      Toplam ücret, aylık ödemelerin zamanla birikmesiyle oluştuğu için müşterinin kalış süresi arttıkça toplam harcamanın artması matematiksel bir gerekliliktir.
MonthlyCharges & TotalCharges (0.65): Beklenen orta-güçlü ilişki. 
      Aylık paket fiyatı ne kadar yüksekse, toplam biriken ücret de o oranda hızlı artar. Ancak süre toplam ücret üzerinde paradan daha belirleyici bir faktör olduğu için korelasyon 0.83 kadar yüksek değildir.
''')

# 9c) tenure vs TotalCharges — scatter plot

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['tenure'], df['TotalCharges'], alpha=0.5, s=30, color='steelblue', edgecolors='navy', linewidth=0.5)

z = np.polyfit(df['tenure'], df['TotalCharges'], 1)
p = np.poly1d(z)
ax.plot(df['tenure'].sort_values(), p(df['tenure'].sort_values()), color='red', linewidth=2.5, label=f'Eğilim (r={correlation_matrix.loc["tenure", "TotalCharges"]:.3f})')

ax.set_xlabel('Tenure (Ay)', fontsize=12, fontweight='bold')
ax.set_ylabel('Total Charges ($)', fontsize=12, fontweight='bold')
ax.set_title('Tenure vs Total Charges Scatter Plot', fontsize=13, fontweight='bold', pad=15)
ax.legend(fontsize=10)
ax.grid(alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()


# 9d) MonthlyCharges vs TotalCharges — scatter plot

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['MonthlyCharges'], df['TotalCharges'], alpha=0.5, s=30, color='seagreen', edgecolors='darkgreen', linewidth=0.5)

z = np.polyfit(df['MonthlyCharges'], df['TotalCharges'], 1)
p = np.poly1d(z)
ax.plot(df['MonthlyCharges'].sort_values(), p(df['MonthlyCharges'].sort_values()), color='red', linewidth=2.5, label=f'Eğilim (r={correlation_matrix.loc["MonthlyCharges", "TotalCharges"]:.3f})')

ax.set_xlabel('Monthly Charges ($)', fontsize=12, fontweight='bold')
ax.set_ylabel('Total Charges ($)', fontsize=12, fontweight='bold')
ax.set_title('Monthly Charges vs Total Charges Scatter Plot', fontsize=13, fontweight='bold', pad=15)
ax.legend(fontsize=10)
ax.grid(alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

```

## 📦 **Aykırı Değer Nedir?**

<div style="background: linear-gradient(135deg, #1a1a2e, #16213e); border-left: 6px solid #ff9a9e; padding: 16px 22px; border-radius: 10px; margin: 16px 0;">
  <strong style="color: #ff9a9e; font-size: 14px;">📦 Aykırı Değer (Outlier)</strong>

    özellikle doğrusal regresyon ve KNN gibi algoritmalar için zararlıdır.
  </p>

## 📐 **IQR Yöntemi**

$$\text{IQR} = Q3 - Q1$$

$$\text{Alt Sınır} = Q1 - 1.5 \times \text{IQR} \qquad \text{Üst Sınır} = Q3 + 1.5 \times \text{IQR}$$

| Adım | İşlem | Kod |
|:---|:---|:---|
| **1** | Çeyreklik hesapla | `Q1 = df[col].quantile(0.25)` |
| **2** | IQR hesapla | `IQR = Q3 - Q1` |
| **3** | Sınırları belirle | `alt = Q1 - 1.5*IQR` |
| **4** | Aykırıları tespit et | `df[(df[col] < alt) \| (df[col] > üst)]` |

<div style="display: grid; grid-template-columns: 1fr 1fr; gap: 15px; margin: 16px 0 24px 0;">
  <div style="background: linear-gradient(135deg, #1a1a2e, #16213e); border-top: 4px solid #69f0ae; padding: 14px 18px; border-radius: 10px; color: white;">
    <strong style="color: #69f0ae;">✅ Telco'da Beklenti</strong>
    <div style="font-size: 13px; color: #b2dfdb; margin-top: 8px; line-height: 1.8;">
      <div>• <code>tenure</code>: Çok az/yok (0–72 bant sınırlı)</div>
      <div>• <code>MonthlyCharges</code>: Bazı aykırı yüksek paket fiyatları</div>
      <div>• <code>TotalCharges</code>: Uzun süreli yüksek ödemeler aykırı görünebilir</div>
    </div>
  </div>
  <div style="background: linear-gradient(135deg, #1a1a2e, #16213e); border-top: 4px solid #f9ca24; padding: 14px 18px; border-radius: 10px; color: white;">

    <div style="font-size: 13px; color: #ffe0b2; margin-top: 8px; line-height: 1.8;">
      <div>• Aykırı değer her zaman <strong>hatalı</strong> veri değildir</div>
      <div>• Gerçek müşteri davranışı da olabilir</div>
      <div>• <strong>Kaldırma</strong> veya <strong>dönüşüm</strong> kararı verime göre verilir</div>
    </div>
  </div>

> ### 🎯 **Göreviniz:**
> **11a)** Her sayısal değişken (`tenure`, `MonthlyCharges`, `TotalCharges`) için **IQR yöntemiyle** alt ve üst sınırları hesaplayın, aykırı değer sayısını ve oranını (%) yazdırın.
>
> **11b)** Üç sayısal değişken için yan yana **boxplot** çizin. Aykırı değer noktalarını görsel olarak işaretleyin (`flierprops` parametresi ile farklı renk).
>
> **11c)** `MonthlyCharges` için aykırı değerleri filtreleyin ve bu müşterilerin `Contract` türü dağılımına bakın — çoğunlukla hangi contract tipinde?
>
> **11d)** Aykırı değer temizleme için IQR ile kırpma (`clipping`) fonksiyonu yazın. Ancak **uygulamayın** — sadece fonksiyon tanımlayın ve karar notunu ekleyin: "Bu veri setinde … olduğu için aykırı değerler korunmuştur."

```python
# ============================================================
# ADIM 11 — Aykırı Değer Analizi (Outlier Detection)
# ============================================================

# 11a) IQR yöntemi ile her sayısal değişkende aykırı değer tespiti
print("\n" + "="*100)
print("11A) IQR YÖNTEMİ İLE AYKIRI DEĞER TESPİTİ")
print("="*100)

# Sadece sürekli sayısal değişkenleri analiz edelim
outlier_columns = ['tenure', 'MonthlyCharges', 'TotalCharges']
outlier_summary = []

for col in outlier_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_count = len(outliers)
    outlier_pct = (outlier_count / len(df)) * 100
    
    print(f"\n📊 {col}:")
    print(f"   Q1 (25%): {q1:.2f}")
    print(f"   Q3 (75%): {q3:.2f}")
    print(f"   IQR: {iqr:.2f}")
    print(f"   Alt Sınır: {lower_bound:.2f}")
    print(f"   Üst Sınır: {upper_bound:.2f}")
    print(f"   Aykırı Değer Sayısı: {outlier_count} (%{outlier_pct:.2f})")
    
    outlier_summary.append({
        'Variable': col,
        'Q1': q1,
        'Q3': q3,
        'IQR': iqr,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outlier Count': outlier_count,
        'Outlier %': outlier_pct
    })

outlier_df = pd.DataFrame(outlier_summary)



# 11b) tenure, MonthlyCharges, TotalCharges için boxplot (yan yana)
print("\n" + "="*100)
print("11B) AYKIRI DEĞER GÖRSELLEŞTORMESI (BOXPLOT)")
print("="*100)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for idx, col in enumerate(outlier_columns):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Boxplot
    bp = axes[idx].boxplot(df[col], vert=True, patch_artist=True)
    bp['boxes'][0].set_facecolor('#3498db')
    bp['medians'][0].set_color('red')
    bp['medians'][0].set_linewidth(2)
    
    # Aykırı değerleri kırmızı ile işaretle
    outliers_mask = (df[col] < lower_bound) | (df[col] > upper_bound)
    outlier_values = df[outliers_mask][col]
    axes[idx].scatter([1]*len(outlier_values), outlier_values, color='red', s=50, alpha=0.6, zorder=3, label='Outliers')
    
    axes[idx].set_ylabel(col, fontsize=11, fontweight='bold')
    axes[idx].set_title(f'{col} Boxplot (Aykırı Değerler İşaretli)', fontsize=12, fontweight='bold')
    axes[idx].grid(axis='y', alpha=0.3, linestyle='--')
    if len(outlier_values) > 0:
        axes[idx].legend()
    axes[idx].set_xticklabels([col])

plt.tight_layout()
plt.show()

# 11c) MonthlyCharges aykırı değerlerine sahip müşterilerin Contract dağılımı
print("\n" + "="*100)
print("11C) MONTHLYCHARGESİ AYKIRI OLAN MÜŞTERİLERİN CONTRACT DAĞILIMI")
print("="*100)

col = 'MonthlyCharges'
q1 = df[col].quantile(0.25)
q3 = df[col].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Aykırı değerleri belirle
outliers_monthly = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
non_outliers_monthly = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print(f"\n📊 MonthlyCharges Aykırı Değer Analizi:")
print(f"   Toplam Müşteri Sayısı: {len(df)}")
print(f"   Aykırı Müşteri Sayısı: {len(outliers_monthly)}")
print(f"   Normal Müşteri Sayısı: {len(non_outliers_monthly)}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Aykırı değerlere sahip müşterilerin Contract dağılımı
outlier_contract = outliers_monthly['Contract'].value_counts()
axes[0].barh(outlier_contract.index, outlier_contract.values, color='#e74c3c')
axes[0].set_xlabel('Count', fontsize=11)
axes[0].set_title('MonthlyCharges AYKIRI Müşterilerin Contract Dağılımı', fontsize=12, fontweight='bold')
axes[0].grid(axis='x', alpha=0.3)

# Normal değerlere sahip müşterilerin Contract dağılımı
normal_contract = non_outliers_monthly['Contract'].value_counts()
axes[1].barh(normal_contract.index, normal_contract.values, color='#2ecc71')
axes[1].set_xlabel('Count', fontsize=11)
axes[1].set_title('MonthlyCharges NORMAL Müşterilerin Contract Dağılımı', fontsize=12, fontweight='bold')
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n🔍 YORUM:")
for contract_type in df['Contract'].unique():
    outlier_in_contract = len(outliers_monthly[outliers_monthly['Contract'] == contract_type])
    total_in_contract = len(df[df['Contract'] == contract_type])
    pct = (outlier_in_contract / total_in_contract * 100) if total_in_contract > 0 else 0
    print(f"   • {contract_type}: {outlier_in_contract}/{total_in_contract} aykırı (%{pct:.1f})")

# 11d) IQR ile aykırı değer kırpma (clipping) fonksiyonunu tanımlayın
```
