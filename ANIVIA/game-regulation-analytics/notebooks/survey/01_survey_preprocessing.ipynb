```python
import pandas as pd
import gspread
import toml
import os
```

```python
# 1. Streamlit şifre kasamızı Jupyter Notebook'ta açıyoruz 
with open("../.streamlit/secrets.toml", "r", encoding="utf-8") as f:
    secrets = toml.load(f)

# 2. Kasadaki bilgilerle Google Cloud'a (GSheets) şifreli giriş yapıyoruz
gc = gspread.service_account_from_dict(secrets["connections"]["gsheets"])

# 3. Tablomuzu ID'si ile bulup canlı açıyoruz
# ID'yi kodun içinden değil, şifre kasasından çekiyoruz
sheet_id = secrets["connections"]["gsheets"]["sheet_id_veli"]
sheet = gc.open_by_key(sheet_id)

# 4. Sayfadaki tüm veriyi çekip anında Pandas DataFrame'ine çeviriyoruz
# Not: get_all_records() hata verebilir (başlıklar benzersiz değilse), bu yüzden get_all_values() kullanıyoruz
data = sheet.sheet1.get_all_values()
df_anket = pd.DataFrame(data[1:], columns=data[0])

print(f"✅ Canlı bağlantı başarılı! Toplam {len(df_anket)} veli verisi anlık olarak çekildi.")
display(df_anket.head())
```

```python
# Boş sütunları tespit et ve sil
bos_sutunlar = df_anket.columns[df_anket.replace(r'^\s*$', pd.NA, regex=True).isna().all()]
print(f"Silinecek boş sütun sayısı: {len(bos_sutunlar)}")

df_anket = df_anket.drop(columns=bos_sutunlar)

print(f"Kalan sütun sayısı: {df_anket.shape[1]}")
display(df_anket.head())
```

```python
df = df_anket.copy()
df.columns
```

```python
# SİSTEM RAPORU: Google Forms'dan gelen 39 adet ham sütun, 
# ML modellerine uygun 'snake_case' formatına çevriliyor.

yeni_kolonlar = [
    "zaman_damgasi",
    "anket_hedefi",                     # Bu anketi kimin adına dolduruyorsunuz?
    
    # --- OYUNCU (KENDİ ADINA) YANITLARI ---
    "oyuncu_cinsiyet",                  # Cinsiyetiniz?
    "oyuncu_yas",                       # Yaşınız?
    "oyuncu_gunluk_sure",               # Oyun oynadığınız günlerde süre...
    "oyuncu_cihaz",                     # EN SIK kullandığınız cihaz...
    "oyuncu_oyun_turu",                 # En çok vakit geçirdiğiniz oyun türü...
    "oyuncu_harcama_butcesi",           # Oyun içi harcamalara ayrılan bütçe...
    "oyuncu_harcama_etkisi",            # Oyun içi harcamalarınız bütçenizi...
    "oyuncu_maglubiyet_tepkisi",        # Rekabetçi bir maçı kaybettiğinizde...
    "oyuncu_motivasyon",                # EN TEMEL motivasyon nedir?
    "oyuncu_stres_etkisi",              # Genel stres seviyenizi nasıl etkiler?
    "oyuncu_uyku_feragati",             # Uykusuz kaldığınız ne sıklıkla olur?
    "oyuncu_ruh_hali",                  # Gerçek hayattaki ruh halinizi...
    "oyuncu_pegi_dikkat",               # Yaş sınırı (PEGI) uyarılarına...
    "oyuncu_platform_kisitlama_gorus",  # Erişim kısıtlanması ihtimalini...

    # --- VELİ VE ÇOCUK YANITLARI ---
    "veli_cinsiyet",                    # Cinsiyetiniz? (Veli formuna girenler için)
    "veli_yas",                         # Yaşınız? (Veli için)
    "cocuk_cinsiyet",                   # Çocuğunuzun Cinsiyeti?
    "cocuk_yas",                        # Çocuğunuzun yaşı?
    "cocuk_cihaz",                      # EN SIK kullandığı cihaz...
    "veli_ebeveyn_denetimi",            # Ebeveyn Denetimi kullanıyor musunuz?
    "cocuk_oyun_turu",                  # En çok vakit geçirdiği oyun türü...
    "cocuk_gunluk_sure",                # Genellikle toplam ne kadar süre...
    "veli_izin_verilen_harcama",        # İzin verilen oyun içi harcama...
    "veli_harcama_guven_etkisi",        # Bütçenizi ve aranızdaki güveni...
    "cocuk_kapatma_tepkisi",            # "Oyun süren bitti" dediğinizde tepkisi...
    "cocuk_stres_gozlemi",              # Genel stres ve kaygı seviyesi...
    "cocuk_uyku_gozlemi",               # Uyku düzeninin bozulduğunu...
    "cocuk_ruh_hali_gozlemi",           # Gerçek hayattaki ruh halini...
    "veli_pegi_kontrol",                # Yaş sınırı etiketlerini kontrol...
    "veli_platform_kisitlama_gorus",    # Erişim kısıtlanması ihtimalini (Veli)...

    # --- DİKKAT VE ORTAK SORULAR ---
    "dikkat_olcumu_1",                  # 'Mavi' rengini işaretleyiniz.
    "dikkat_olcumu_2",                  # 'Mavi' rengini işaretleyiniz. (Tekrarlı soru)
    "oyuncu_harcama_amaci",             # Harcama YAPIYORSANIZ en temel amacı...
    "cocuk_haftalik_oyun_gunu",         # Çocuğunuz haftanın kaç günü oyun oynar?
    "oyuncu_haftalik_oyun_gunu",        # Haftanın kaç günü oyun oynarsınız?
    "veli_ebeveyn_onay_gorus",        # Ebeveyn onayı getirmeli midir?
    "oyuncu_ebeveyn_onay_gorus"           # Ebeveyn onayı getirmeli midir? (Tekrarlı soru)
]

# Pandas'ın kolon isimlerini doğrudan kusursuz listemizle değiştiriyoruz
df.columns = yeni_kolonlar

# Başarı kontrolü için ilk 3 satırı ekrana basalım
print("✅ Kolon isimleri başarıyla 'snake_case' yapısına dönüştürüldü!")
display(df.head(3))
```

```python
import numpy as np

# 1. ÖNCE TEMİZLİK: Google Sheets'in sahte boşluklarını ("") gerçek NaN değerine çeviriyoruz.
# regex=True sayesinde içinde sadece görünmez bir boşluk (" ") olan hatalı hücreler de yakalanır.
df = df.replace(r'^\s*$', np.nan, regex=True)

# 2. BÖLME İŞLEMİ VE SİLME (Şimdi dropna sorunsuz çalışacak)

# KOL 1: OYUNCULAR
df_oyuncular = df[df['anket_hedefi'] == '18 Yaş Üzeriyim - Kendim İçin'].copy()
df_oyuncular = df_oyuncular.dropna(axis=1, how='all')

# KOL 2: VELİLER
df_veliler = df[df['anket_hedefi'] == '18 Yaş Üzeriyim - Çocuğumun alışkanlıkları için'].copy()
df_veliler = df_veliler.dropna(axis=1, how='all')

# SİSTEM RAPORU
print(f"✅ BÖLÜNME VE SÜTUN TEMİZLİĞİ BAŞARILI!")
print(f"🎮 Oyuncu Tablosu: {len(df_oyuncular)} satır, {len(df_oyuncular.columns)} sütun")
print(f"👨‍👩‍👧 Veli Tablosu: {len(df_veliler)} satır, {len(df_veliler.columns)} sütun")
```

```python
# Güvenlik filtresi fonksiyonumuz
def guvenlik_testi_uygula(veri):
    # İçinde 'dikkat_olcumu' geçen hayatta kalmış sütunları bul
    dikkat_sutunlari = [col for col in veri.columns if 'dikkat_olcumu' in col]
    
    if dikkat_sutunlari:
        # Sütunlardaki NaN değerleri boş string yapıp birleştiriyoruz
        # 'mavi' kelimesini (büyük/küçük harf duyarsız) arıyoruz
        sart = veri[dikkat_sutunlari].fillna('').astype(str).apply(
            lambda x: 'mavi' in ' '.join(x).lower(), axis=1
        )
        
        # Sadece testi geçenleri tut
        temiz_veri = veri[sart].copy()
        
        # Test bittiği için artık hiçbir işe yaramayan dikkat sütunlarını çöpe at (Feature Selection)
        temiz_veri = temiz_veri.drop(columns=dikkat_sutunlari)
        
        return temiz_veri
        
    return veri # Eğer dikkat sütunu hiç yoksa veriyi olduğu gibi döndür

# Filtreyi her iki tabloya da uyguluyoruz
df_oyuncular = guvenlik_testi_uygula(df_oyuncular)
df_veliler = guvenlik_testi_uygula(df_veliler)

# Son durumu raporlayalım
print("🛡️ GÜVENLİK FİLTRESİ UYGULANDI (Rastgele dolduranlar elendi ve dikkat sütunları silindi)")
print(f"🎮 Kalan Gerçek Oyuncu Sayısı: {len(df_oyuncular)}")
print(f"👨‍👩‍👧 Kalan Gerçek Veli Sayısı: {len(df_veliler)}")
```

```python
# ==========================================
# 🧹 EDA HEDEFLERİ VE VERİ TEMİZLİĞİ
# ==========================================

# ------------------------------------------
# 1. OYUNCULAR (df_oyuncular) İŞLEMLERİ
# ------------------------------------------

# Hedef 1: Gereksiz sütunları sil
silinecekler = ['zaman_damgasi', 'anket_hedefi']
df_oyuncular = df_oyuncular.drop(columns=silinecekler, errors='ignore')

# Hedef 2: 'Belirtmek İstemiyorum' diyenleri oyuncu_cinsiyet sütununun Mod'u ile doldur
if 'oyuncu_cinsiyet' in df_oyuncular.columns:
    # "Belirtmek İstemiyorum" haricindeki en çok geçen değeri (mod) bul
    oyuncu_mod = df_oyuncular[df_oyuncular['oyuncu_cinsiyet'] != 'Belirtmek İstemiyorum']['oyuncu_cinsiyet'].mode()[0]
    # Değişimi uygula
    df_oyuncular['oyuncu_cinsiyet'] = df_oyuncular['oyuncu_cinsiyet'].replace('Belirtmek İstemiyorum', oyuncu_mod)

# Hedef 3: oyuncu_harcama_amaci sütunundaki boş (NaN) değerleri doldur
if 'oyuncu_harcama_amaci' in df_oyuncular.columns:
    df_oyuncular['oyuncu_harcama_amaci'] = df_oyuncular['oyuncu_harcama_amaci'].fillna("Oyun içi harcama yapmam.")


# ------------------------------------------
# 2. VELİLER (df_veliler) İŞLEMLERİ
# ------------------------------------------

# Hedef 1: Gereksiz sütunları sil
df_veliler = df_veliler.drop(columns=silinecekler, errors='ignore')

# Hedef 2: veli_cinsiyet için 'Belirtmek İstemiyorum' olanları Mod ile doldur
if 'veli_cinsiyet' in df_veliler.columns:
    veli_mod = df_veliler[df_veliler['veli_cinsiyet'] != 'Belirtmek İstemiyorum']['veli_cinsiyet'].mode()[0]
    df_veliler['veli_cinsiyet'] = df_veliler['veli_cinsiyet'].replace('Belirtmek İstemiyorum', veli_mod)

# Hedef 3: cocuk_cinsiyet için 'Belirtmek İstemiyorum' olanları Mod ile doldur
if 'cocuk_cinsiyet' in df_veliler.columns:
    cocuk_mod = df_veliler[df_veliler['cocuk_cinsiyet'] != 'Belirtmek İstemiyorum']['cocuk_cinsiyet'].mode()[0]
    df_veliler['cocuk_cinsiyet'] = df_veliler['cocuk_cinsiyet'].replace('Belirtmek İstemiyorum', cocuk_mod)

# SİSTEM RAPORU
print("✅ EDA Temizlik Hedefleri Başarıyla Tamamlandı!")
print("-" * 40)
print(f"🎮 Oyuncu Tablosu Kalan Sütun Sayısı: {len(df_oyuncular.columns)}")
print(f"👨‍👩‍👧 Veli Tablosu Kalan Sütun Sayısı: {len(df_veliler.columns)}")
```

```python
def ardisik_tekrarlari_sil(df):
    # df.shift(1) veriyi bir satır aşağı kaydırır. 
    # Böylece her satırı doğrudan bir üstündeki satırla kıyaslamış oluruz.
    # all(axis=1) ise "Tüm sütunlar birebir aynı mı?" diye bakar.
    ardisik_mi = (df == df.shift(1)).all(axis=1)
    
    silinen_sayisi = ardisik_mi.sum()
    
    # Ardışık olmayan (True dönmeyen) temiz satırları tutuyoruz
    temiz_df = df[~ardisik_mi].copy()
    
    print(f"🧹 Ardışık Tekrar Kontrolü: {silinen_sayisi} adet çift tıklama (ardışık) kaydı silindi.")
    return temiz_df

# Fonksiyonu tablolarımıza uyguluyoruz
df_oyuncular = ardisik_tekrarlari_sil(df_oyuncular)
df_veliler = ardisik_tekrarlari_sil(df_veliler)
```

```python
# Kayıt yapacağımız klasörün var olduğundan emin olalım (Yosa otomatik açar)
hedef_klasor = '../data/processed'
os.makedirs(hedef_klasor, exist_ok=True)

# Verileri doğrudan ait oldukları işlenmiş veri klasörüne kaydediyoruz
df_oyuncular.to_csv(f'{hedef_klasor}/oyuncular_temiz.csv', index=False)
df_veliler.to_csv(f'{hedef_klasor}/veliler_temiz.csv', index=False)

print("💾 Veriler başarıyla 'data/processed' klasörüne kaydedildi!")
```
