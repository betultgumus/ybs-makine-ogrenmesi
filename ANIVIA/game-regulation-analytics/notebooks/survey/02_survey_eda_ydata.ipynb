```python
import pandas as pd
from ydata_profiling import ProfileReport

# 0. VERİYİ DOĞRU YERDEN OKUMA (Cookiecutter Mimarisi)
print("📂 Temiz veriler 'data/processed' klasöründen yükleniyor...")
df_oyuncular = pd.read_csv('../../data/processed/oyuncular_temiz.csv')
df_veliler = pd.read_csv('../../data/processed/veliler_temiz.csv')

# SİSTEM RAPORU: ydata-profiling ile Otomatik Keşifsel Veri Analizi (EDA) başlatılıyor...

# 1. OYUNCULAR İÇİN RAPOR OLUŞTURMA
print("🎮 Oyuncu verisi analiz ediliyor... (Bu işlem verinin boyutuna göre 15-30 saniye sürebilir)")
oyuncu_raporu = ProfileReport(df_oyuncular, title="Oyuncu EDA Raporu", explorative=True)
oyuncu_raporu.to_file("../../reports/oyuncular_eda_raporu.html")

# 2. VELİLER İÇİN RAPOR OLUŞTURMA
print("👨‍👩‍👧 Veli verisi analiz ediliyor...")
veli_raporu = ProfileReport(df_veliler, title="Veli EDA Raporu", explorative=True)
veli_raporu.to_file("../../reports/veliler_eda_raporu.html")

print("✅ Raporlar başarıyla oluşturuldu ve klasöre kaydedildi!")
```
